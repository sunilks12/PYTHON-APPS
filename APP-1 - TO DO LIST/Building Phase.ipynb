# while and match

```python
todos = []
while True:
    user_action = input("Enter add or show or edit or exit: ")
    user_action = user_action.strip()
    
    match user_action:
        
        case 'add':
            todo = input("Enter todo: ")
            todos.append(todo)
            
        case 'show' |'display':
            for i in todos:
                i=i.title()
                print(i)
        case 'edit':
            number=int(input("Enter which number todo to be replaced: "))
            new_todo=input("Enter new todo: ")
            todos[number-1]=new_todo
            
        case 'exit':
            break
        case whatever:
            print("Hey you have entered the wrong command!")
print('Thanks')           
```
